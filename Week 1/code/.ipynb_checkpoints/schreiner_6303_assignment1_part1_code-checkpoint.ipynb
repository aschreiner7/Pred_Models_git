### Task 6 - Perform basic correlation analysis a correlation matrix) among the attributes. Perform both numeric and visual correlation outputs.

```python
## AJS - #Visual correlation output shows that there are some clear linear patterns between the fields

sns.set(style='whitegrid', context='notebook')
cols = ['Mileage (mpg)', 'Capacity (gal)', 'Length (inches)', 'Wheel Base (inches)', 'Width (inches)', 'Luggage (cu. ft.)', 'Turning Circle (feet)']


sns.pairplot(auto[cols], height=2.5)
plt.tight_layout()
# plt.savefig('../figures/scatter.png', dpi=300)
plt.show()
```

```python
## AJS - Numeric correlation matrix on Mileage, shows that all the size variables have a negative correlation to the MPG
##       as the size of the gas tank increases, the MPG will be lower. This is the most correlated example with a value of -.79
##       It is also worth noting that the values have high correlations between themselves (which is intuitive and expected). 

cm = np.corrcoef(auto[cols].values.T)

#initialize seaborn library
sns.set_theme(font_scale=1, rc = {'figure.figsize':(9,6)})

#create seaborn object
hm = sns.heatmap(cm, 
              cbar=True,
              annot=True, 
              square=True,
              fmt='.2f',
              yticklabels=cols,
              xticklabels=cols)
                

plt.tight_layout()
# plt.savefig('../figures/corr_mat.png', dpi=300)
plt.show()
```

### Task 7 - Using the new converted dataset- run a PCA. How much variation is captured in 3 Principle Components?

```python
## AJS - creating a limited data frame and including the new weight value, I opted not to include the categorical dummies this time

auto_numerics_only = auto.select_dtypes(include=np.number)
auto_numerics_only

#special thanks to https://stackoverflow.com/questions/25039626/how-do-i-find-numeric-columns-in-pandas
```

```python
## AJS - I believe this is the function referenced by the manual code from week 1, it was used in both week 2 code files. 
##       I also found the referenced stackoverflow page below to be helpful. 

scaled_features = StandardScaler().fit_transform(auto_numerics_only.values)
scaled_features_df = pd.DataFrame(scaled_features, index=auto_numerics_only.index, columns=auto_numerics_only.columns)
scaled_features_df.head()

#special thanks to https://stackoverflow.com/questions/35723472/how-to-use-sklearn-fit-transform-with-pandas-and-return-dataframe-instead-of-num 
```

```python
## AJS - for some reason, this took a while for me to pull off. I see it is not complicated once I got it working, 
##       but it took some time and digging around. The PCA from Week 1 is nested in with a logistic regression code block 
##       with training and test sets which this simple example does not have. The directions suggest to run it with 3 
##       principle components, but it would be cool to see the code to do a scree plot, means to review the eigenvalues, 
##       and create a visual of the outputs. I have done this in R, just need to learn how to do it in Python now. :-)

pca = PCA(n_components=3)

#fit and transform the feature space
features_pca = pca.fit_transform(scaled_features)

#create a dataframe for the output
pca_df = pd.DataFrame(data = features_pca,
                           columns = ['principal component 1', 'principal component 2','principal component 3' ])
pca_df.head()
```

### Task 10 - Perform a cross-tabulation of the two "fuel type" variables versus the two "Drive type" variables.

```python
## AJS - looks like premium gas has an almost equal amount of front and rear drivetrain observations compared to regular
##       this is intuitive given the larger number of sports cars that are rear wheel drive. Suprising that AWD or 4WD is not listed.

#Crosstab using Pandas
auto_cross=pd.crosstab(auto['Fuel Type'], auto['Drive Type'],colnames=['Drive Type'],rownames=['Fuel Type'])
print(auto_cross)
```
